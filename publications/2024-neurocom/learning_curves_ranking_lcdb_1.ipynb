```python
import matplotlib.pyplot as plt
import pandas as pd

from deephyper.analysis._matplotlib import figure_size, update_matplotlib_rc
update_matplotlib_rc()
figsize = figure_size(252 * 1.8, 1.0)

from lcdb.analysis.plot import plot_learning_curves
from util import * # utility functions for this evaluation
```

Download the data from https://surfdrive.surf.nl/files/index.php/s/VJtGuQaDw2otdZh/download.

```python
df = pd.read_csv("database-accuracy.csv")
df
```

```python
df_mean_curves = get_mean_curves(df)
```

```python
df_mean_curves
```

```python
def process_openmlid(df_mean_curves, openmlid):
    temp = df_mean_curves[df_mean_curves['openmlid'] == openmlid]
    sizes = []
    mean_errorrates = []
    for i in range(len(temp)):
        row = temp.iloc[i]
        sizes.append(row.sizes)
        mean_errorrates.append(row.mean_errorrates)

    sizes_array = sizes
    mean_errorrates_array = mean_errorrates
    return sizes_array, mean_errorrates_array
```

```python
len(df_mean_curves['openmlid'].unique())/3
```

```python
for openmlid in df_mean_curves["openmlid"].unique():
    print(f"Drawing openml.{openmlid}")
    fig, ax = plt.subplots(figsize=figsize)
    sizes_array, mean_errorrates_array = process_openmlid(df_mean_curves, openmlid)
    fig, ax = plot_learning_curves(sizes_array, mean_errorrates_array, ax=ax)
    ax.set_xlabel(f"Number of Samples")
    ax.set_ylabel(f"Expected Validation Error Rate")
    ax.set_xscale("log")
    ax.set_yscale("log")
    plt.savefig(f"figures/openmlid-{openmlid}.jpg", dpi=300, bbox_inches="tight")
    plt.close()
```
